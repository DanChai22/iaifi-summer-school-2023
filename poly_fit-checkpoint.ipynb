```python
%load_ext autoreload
%autoreload 2

import sys, os
from IPython.display import display


import numpy as np
import scipy as sp
import pandas as pd
import numpy.random as rand
import numpy.linalg as la

from scipy.special import eval_legendre

import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import itertools as it

import pickle
import time
```

```python
def legendre_fit(x, y, Np):
    '''
    Performs polynomial regression with a Legendre polynomial basis.
    Np: Number of basis elements to use. Polynomial is order Np-1.
    Returns: Coefficients of fit polynomial.
    '''
    
    X = np.zeros([len(y), Np])
    
    for n in range(Np):
        X[:, n] = eval_legendre(n, x)/ np.sqrt(2 / (2*n+1))
        # unormalized version
        # X[:, n] = eval_legendre(n, x)
        
    
    w = la.solve(X.T@X + 1e-8*np.identity(Np), X.T@y)
    
    return w

def eval_fit(x, w):
    ''' 
    Evaluates Legendre polynomial at given x coordinates with coefficients w.
    '''
    
    y = np.zeros_like(x)
    for n in range(len(w)):
        y += w[n]*eval_legendre(n, x)/ np.sqrt(2 / (2*n+1))
        # unormalized version
        # y += w[n]*eval_legendre(n, x)
        
    return y
        

def plot_fit(M, Np):
    '''
    M: Number of training data points.
    Np: Number of fit parameters.
    '''
    
    rand.seed(15)
    
    sns.set(context='talk', font_scale=1.0, color_codes=True, palette='deep', style='ticks', 
        rc={'mathtext.fontset': 'cm', 'xtick.direction': 'in','ytick.direction': 'in',
            'axes.linewidth': 1.5, 'figure.dpi':100, 'text.usetex':False})

    fig = plt.figure(figsize=(8, 5))

    ax = fig.add_subplot(1, 1, 1)

    func = lambda x:  0.4*np.sin(2*np.pi*x)+ 0.5*x + x**2 -0.5

    x = np.linspace(-1.0, 1.0, 1000)
    ax.plot(x, func(x), 'k--', lw=1.5, label='Ground Truth')

    
    # x values are equally spaced plus some random noise and then clipped to fit the interval
    x = np.linspace(-1.0, 1.0, M) + rand.normal(scale=0.1, size=M)
    x = np.clip(x, -1.0, 1.0)

    y = func(x) + rand.normal(scale=0.1, size=M)
    ax.plot(x, y, 'ob', ms=8.0,  label='Training Data')

    w = legendre_fit(x, y, Np)

    x = np.linspace(-1.0, 1.0, 1000)
    p = eval_fit(x, w)
    ax.plot(x, p, '-', color=sns.color_palette('deep')[1], label="Model")

    Mtest = 25
    xtest = np.linspace(-1.0, 1.0, Mtest) + rand.normal(scale=0.1, size=Mtest)
    xtest = np.clip(xtest, -1.0, 1.0)

    ytest = func(xtest) + rand.normal(scale=0.1, size=Mtest)
    ax.plot(xtest, ytest, 'sr', ms=8.0,  label='Test Data')

    # ax.hlines(0, xmin=-1.1, xmax=1.1, color='k', ls='-')

    ax.set_xlim(-1.0, 1.0)
    ax.set_ylim(-1.5, 2.0)

    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylabel("Labels $y$")
    ax.set_xlabel("Features $x$")

    ax.legend(frameon=True, edgecolor='k', fancybox=False, loc='upper left', fontsize='small', framealpha=1.0)

    # plt.savefig("poly_fit.png", dpi=300, bbox_inches='tight')


    plt.show()


    
```

# Very Underparameterized

```python
M = 25
Np = 3

plot_fit(M, Np)
```
